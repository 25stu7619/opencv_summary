```python
# Google ColabÏóêÏÑú OpenCVÎ°ú Ïù∏ÌÑ∞ÎÑ∑ Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú Î∞è Ï∂úÎ†•
# Ïù¥ ÏΩîÎìúÎ•º Google ColabÏóêÏÑú Ïã§ÌñâÌïòÏÑ∏Ïöî

# ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò Î∞è ÏûÑÌè¨Ìä∏
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

print("OpenCV Î≤ÑÏ†Ñ:", cv2.__version__)

# ============================================
# Î∞©Î≤ï 1: urllibÏùÑ ÏÇ¨Ïö©Ìïú Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú
# ============================================
def download_image_urllib(url, save_path='downloaded_image.jpg'):
    """
    urllibÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ïù¥ÎØ∏ÏßÄÎ•º Îã§Ïö¥Î°úÎìúÌïòÎäî Ìï®Ïàò

    Args:
        url (str): Ïù¥ÎØ∏ÏßÄ URL
        save_path (str): Ï†ÄÏû•Ìï† ÌååÏùº Í≤ΩÎ°ú

    Returns:
        numpy.ndarray: OpenCV Ïù¥ÎØ∏ÏßÄ Î∞∞Ïó¥
    """
    try:
        # Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú
        urllib.request.urlretrieve(url, save_path)
        print(f"‚úì Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú ÏôÑÎ£å: {save_path}")

        # OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞
        img = cv2.imread(save_path)

        if img is None:
            print("‚úó Ïù¥ÎØ∏ÏßÄÎ•º ÏùΩÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")
            return None

        return img

    except Exception as e:
        print(f"‚úó Îã§Ïö¥Î°úÎìú Ïã§Ìå®: {e}")
        return None


# ============================================
# Î∞©Î≤ï 2: requestsÎ•º ÏÇ¨Ïö©Ìïú Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú
# ============================================
def download_image_requests(url, save_path='downloaded_image.jpg'):
    """
    requestsÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Ïù¥ÎØ∏ÏßÄÎ•º Îã§Ïö¥Î°úÎìúÌïòÎäî Ìï®Ïàò

    Args:
        url (str): Ïù¥ÎØ∏ÏßÄ URL
        save_path (str): Ï†ÄÏû•Ìï† ÌååÏùº Í≤ΩÎ°ú

    Returns:
        numpy.ndarray: OpenCV Ïù¥ÎØ∏ÏßÄ Î∞∞Ïó¥
    """
    try:
        # Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú
        response = requests.get(url)
        response.raise_for_status()  # HTTP ÏóêÎü¨ Ï≤¥ÌÅ¨

        # Î∞îÏù¥Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º ÌååÏùºÎ°ú Ï†ÄÏû•
        with open(save_path, 'wb') as f:
            f.write(response.content)

        print(f"‚úì Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú ÏôÑÎ£å: {save_path}")

        # OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞
        img = cv2.imread(save_path)

        if img is None:
            print("‚úó Ïù¥ÎØ∏ÏßÄÎ•º ÏùΩÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")
            return None

        return img

    except Exception as e:
        print(f"‚úó Îã§Ïö¥Î°úÎìú Ïã§Ìå®: {e}")
        return None


# ============================================
# Î∞©Î≤ï 3: Î©îÎ™®Î¶¨ÏóêÏÑú ÏßÅÏ†ë Ï≤òÎ¶¨ (Ï†ÄÏû• ÏóÜÏù¥)
# ============================================
def download_image_to_memory(url):
    """
    Ïù¥ÎØ∏ÏßÄÎ•º Î©îÎ™®Î¶¨ÏóêÏÑú ÏßÅÏ†ë Ï≤òÎ¶¨ (ÎîîÏä§ÌÅ¨Ïóê Ï†ÄÏû•ÌïòÏßÄ ÏïäÏùå)

    Args:
        url (str): Ïù¥ÎØ∏ÏßÄ URL

    Returns:
        numpy.ndarray: OpenCV Ïù¥ÎØ∏ÏßÄ Î∞∞Ïó¥
    """
    try:
        # Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú
        response = requests.get(url)
        response.raise_for_status()

        # Î∞îÏù¥Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º numpy Î∞∞Ïó¥Î°ú Î≥ÄÌôò
        image_array = np.asarray(bytearray(response.content), dtype=np.uint8)

        # OpenCVÎ°ú ÎîîÏΩîÎî©
        img = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

        if img is None:
            print("‚úó Ïù¥ÎØ∏ÏßÄÎ•º ÎîîÏΩîÎî©Ìï† Ïàò ÏóÜÏäµÎãàÎã§.")
            return None

        print("‚úì Ïù¥ÎØ∏ÏßÄ Î°úÎìú ÏôÑÎ£å (Î©îÎ™®Î¶¨)")
        return img

    except Exception as e:
        print(f"‚úó Îã§Ïö¥Î°úÎìú Ïã§Ìå®: {e}")
        return None


# ============================================
# Ïù¥ÎØ∏ÏßÄ Ï†ïÎ≥¥ Ï∂úÎ†• Ìï®Ïàò
# ============================================
def print_image_info(img, image_name="Ïù¥ÎØ∏ÏßÄ"):
    """
    Ïù¥ÎØ∏ÏßÄÏùò ÏÉÅÏÑ∏ Ï†ïÎ≥¥Î•º Ï∂úÎ†•ÌïòÎäî Ìï®Ïàò

    Args:
        img (numpy.ndarray): OpenCV Ïù¥ÎØ∏ÏßÄ Î∞∞Ïó¥
        image_name (str): Ïù¥ÎØ∏ÏßÄ Ïù¥Î¶Ñ
    """
    if img is None:
        print("Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏóÜÏäµÎãàÎã§.")
        return

    print(f"\n{'='*50}")
    print(f"üìä {image_name} Ï†ïÎ≥¥")
    print(f"{'='*50}")
    print(f"Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞: {img.shape}")
    print(f"ÎÜíÏù¥: {img.shape[0]} ÌîΩÏÖÄ")
    print(f"ÎÑàÎπÑ: {img.shape[1]} ÌîΩÏÖÄ")

    if len(img.shape) == 3:
        print(f"Ï±ÑÎÑê Ïàò: {img.shape[2]} (Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄ)")
    else:
        print(f"Ï±ÑÎÑê Ïàò: 1 (Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄ)")

    print(f"Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ: {img.dtype}")
    print(f"Ï†ÑÏ≤¥ ÌîΩÏÖÄ Ïàò: {img.size:,}")
    print(f"Î©îÎ™®Î¶¨ ÌÅ¨Í∏∞: {img.nbytes / 1024:.2f} KB")
    print(f"ÏµúÏÜå ÌîΩÏÖÄ Í∞í: {img.min()}")
    print(f"ÏµúÎåÄ ÌîΩÏÖÄ Í∞í: {img.max()}")
    print(f"ÌèâÍ∑† ÌîΩÏÖÄ Í∞í: {img.mean():.2f}")
    print(f"{'='*50}\n")


# ============================================
# Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî Ìï®Ïàò (ColabÏö©)
# ============================================
def display_image(img, title="Ïù¥ÎØ∏ÏßÄ", figsize=(10, 8)):
    """
    Google ColabÏóêÏÑú Ïù¥ÎØ∏ÏßÄÎ•º ÏãúÍ∞ÅÌôîÌïòÎäî Ìï®Ïàò

    Args:
        img (numpy.ndarray): OpenCV Ïù¥ÎØ∏ÏßÄ Î∞∞Ïó¥
        title (str): Ïù¥ÎØ∏ÏßÄ Ï†úÎ™©
        figsize (tuple): Í∑∏Î¶º ÌÅ¨Í∏∞
    """
    if img is None:
        print("ÌëúÏãúÌï† Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏóÜÏäµÎãàÎã§.")
        return

    # OpenCVÎäî BGRÎ°ú ÏùΩÏúºÎØÄÎ°ú RGBÎ°ú Î≥ÄÌôò
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=figsize)
    plt.imshow(img_rgb)
    plt.title(title, fontsize=16, fontweight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.show()


# ============================================
# Ïó¨Îü¨ Ïù¥ÎØ∏ÏßÄÎ•º Í≤©ÏûêÎ°ú ÌëúÏãúÌïòÎäî Ìï®Ïàò
# ============================================
def display_images_grid(images, titles, rows=2, cols=2, figsize=(15, 12)):
    """
    Ïó¨Îü¨ Ïù¥ÎØ∏ÏßÄÎ•º Í≤©Ïûê ÌòïÌÉúÎ°ú ÌëúÏãúÌïòÎäî Ìï®Ïàò

    Args:
        images (list): OpenCV Ïù¥ÎØ∏ÏßÄ Î∞∞Ïó¥ Î¶¨Ïä§Ìä∏
        titles (list): Ïù¥ÎØ∏ÏßÄ Ï†úÎ™© Î¶¨Ïä§Ìä∏
        rows (int): Ìñâ Ïàò
        cols (int): Ïó¥ Ïàò
        figsize (tuple): Í∑∏Î¶º ÌÅ¨Í∏∞
    """
    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    axes = axes.flatten()

    for i, (img, title) in enumerate(zip(images, titles)):
        if i >= rows * cols:
            break

        if img is None:
            axes[i].text(0.5, 0.5, 'Ïù¥ÎØ∏ÏßÄ ÏóÜÏùå',
                        ha='center', va='center', fontsize=12)
            axes[i].set_title(title)
            axes[i].axis('off')
            continue

        # BGR to RGB Î≥ÄÌôò
        if len(img.shape) == 3 and img.shape[2] == 3:
            img_display = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        else:
            img_display = img

        axes[i].imshow(img_display, cmap='gray' if len(img.shape) == 2 else None)
        axes[i].set_title(title, fontsize=12, fontweight='bold')
        axes[i].axis('off')

    # Îπà ÏÑúÎ∏åÌîåÎ°Ø Ïà®Í∏∞Í∏∞
    for i in range(len(images), rows * cols):
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()


# ============================================
# Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ ÏòàÏ†ú Ìï®ÏàòÎì§
# ============================================
def process_images(img):
    """
    Îã§ÏñëÌïú Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨Î•º ÏàòÌñâÌïòÎäî Ìï®Ïàò

    Args:
        img (numpy.ndarray): ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ

    Returns:
        dict: Ï≤òÎ¶¨Îêú Ïù¥ÎØ∏ÏßÄÎì§Ïùò ÎîïÏÖîÎÑàÎ¶¨
    """
    processed = {}

    # Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Î≥ÄÌôò
    processed['grayscale'] = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Í∞ÄÏö∞ÏãúÏïà Î∏îÎü¨
    processed['blurred'] = cv2.GaussianBlur(img, (15, 15), 0)

    # Canny Ïó£ÏßÄ Í≤ÄÏ∂ú
    processed['edges'] = cv2.Canny(processed['grayscale'], 100, 200)

    # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ Ï°∞Ï†ï
    height, width = img.shape[:2]
    processed['resized'] = cv2.resize(img, (width//2, height//2))

    # HSV Î≥ÄÌôò
    processed['hsv'] = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # ÌûàÏä§ÌÜ†Í∑∏Îû® ÌèâÌôúÌôî (Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº)
    processed['equalized'] = cv2.equalizeHist(processed['grayscale'])

    return processed


# ============================================
# Î©îÏù∏ Ïã§Ìñâ ÏΩîÎìú
# ============================================
if __name__ == "__main__":
    print("="*60)
    print("üñºÔ∏è  OpenCV Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú Î∞è Ï≤òÎ¶¨ ÌîÑÎ°úÍ∑∏Îû®")
    print("="*60)

    # ÏÇ¨Ïö©ÏûêÎ°úÎ∂ÄÌÑ∞ Ïù¥ÎØ∏ÏßÄ URL ÏûÖÎ†•Î∞õÍ∏∞
    print("\nüìù Ïù¥ÎØ∏ÏßÄ URLÏùÑ ÏûÖÎ†•ÌïòÏÑ∏Ïöî.")
    print("(ÏòàÏãú: https://images.unsplash.com/photo-1514888286974-6c03e2ca1dba?w=800)")
    print("-" * 60)

    # Google ColabÏóêÏÑú ÏÇ¨Ïö©Ïûê ÏûÖÎ†• Î∞õÍ∏∞
    url = input("Ïù¥ÎØ∏ÏßÄ URL: ").strip()

    # URLÏù¥ ÎπÑÏñ¥ÏûàÎäî Í≤ΩÏö∞ Í∏∞Î≥∏ Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïö©
    if not url:
        print("\n‚ö†Ô∏è  URLÏù¥ ÏûÖÎ†•ÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§. Í∏∞Î≥∏ Ïù¥ÎØ∏ÏßÄÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.")
        url = "https://images.unsplash.com/photo-1514888286974-6c03e2ca1dba?w=800"

    print(f"\nüì• Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú Ï§ë...\n")
    print(f"URL: {url}")

    # Î∞©Î≤ï 1: urllib ÏÇ¨Ïö©
    print("\n--- Î∞©Î≤ï 1: urllib ÏÇ¨Ïö© ---")
    img1 = download_image_urllib(url, 'image_urllib.jpg')

    # Î∞©Î≤ï 2: requests ÏÇ¨Ïö©
    print("\n--- Î∞©Î≤ï 2: requests ÏÇ¨Ïö© ---")
    img2 = download_image_requests(url, 'image_requests.jpg')

    # Î∞©Î≤ï 3: Î©îÎ™®Î¶¨ÏóêÏÑú ÏßÅÏ†ë Ï≤òÎ¶¨
    print("\n--- Î∞©Î≤ï 3: Î©îÎ™®Î¶¨ÏóêÏÑú ÏßÅÏ†ë Ï≤òÎ¶¨ ---")
    img3 = download_image_to_memory(url)

    # Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Î°úÎìúÎêòÏóàÎäîÏßÄ ÌôïÏù∏
    if img1 is not None:
        # Ïù¥ÎØ∏ÏßÄ Ï†ïÎ≥¥ Ï∂úÎ†•
        print_image_info(img1, "Îã§Ïö¥Î°úÎìúÎêú Ïù¥ÎØ∏ÏßÄ")

        # Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
        print("üìä ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ ÌëúÏãú")
        display_image(img1, "ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ - Í≥†ÏñëÏù¥")

        # Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ ÏàòÌñâ
        print("\nüîÑ Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ Ï§ë...\n")
        processed_images = process_images(img1)

        # Ï≤òÎ¶¨Îêú Ïù¥ÎØ∏ÏßÄÎì§ ÌëúÏãú
        images_to_display = [
            img1,
            processed_images['grayscale'],
            processed_images['blurred'],
            processed_images['edges'],
        ]

        titles = [
            "ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ",
            "Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº",
            "Í∞ÄÏö∞ÏãúÏïà Î∏îÎü¨",
            "Ïó£ÏßÄ Í≤ÄÏ∂ú",
        ]

        print("üìä Ï≤òÎ¶¨Îêú Ïù¥ÎØ∏ÏßÄÎì§ ÌëúÏãú")
        display_images_grid(images_to_display, titles, rows=2, cols=2)

        # Ï∂îÍ∞Ä Ï†ïÎ≥¥ Ï∂úÎ†•
        print("\n‚úÖ Î™®Îì† Ï≤òÎ¶¨ ÏôÑÎ£å!")
        print(f"Ï≤òÎ¶¨Îêú Ïù¥ÎØ∏ÏßÄ Ïàò: {len(processed_images)}")
        print(f"Ï†ÄÏû•Îêú ÌååÏùº: image_urllib.jpg, image_requests.jpg")

    # ============================================
    # Ïó¨Îü¨ Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
    # ============================================
    print("\n" + "="*60)
    print("üì• Ï∂îÍ∞Ä Ïù¥ÎØ∏ÏßÄÎ•º Îçî Îã§Ïö¥Î°úÎìúÌïòÏãúÍ≤†ÏäµÎãàÍπå?")
    print("="*60 + "\n")

    more_images = input("Ï∂îÍ∞Ä Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú? (y/n): ").strip().lower()

    if more_images == 'y':
        downloaded_images = []
        image_names = []

        # Ï≤´ Î≤àÏß∏ Ïù¥ÎØ∏ÏßÄ Ï∂îÍ∞Ä
        if img1 is not None:
            img1_resized = cv2.resize(img1, (400, 300))
            downloaded_images.append(img1_resized)
            image_names.append("Ïù¥ÎØ∏ÏßÄ 1")

        # Ï∂îÍ∞Ä Ïù¥ÎØ∏ÏßÄ ÏûÖÎ†• Î∞õÍ∏∞ (ÏµúÎåÄ 3Í∞ú Îçî)
        for i in range(2, 5):
            url_input = input(f"\nÏù¥ÎØ∏ÏßÄ {i} URL (ÏóîÌÑ∞Î•º ÎàÑÎ•¥Î©¥ Í±¥ÎÑàÎõ∞Í∏∞): ").strip()

            if not url_input:
                break

            print(f"Îã§Ïö¥Î°úÎìú Ï§ë...")
            img = download_image_to_memory(url_input)

            if img is not None:
                img_resized = cv2.resize(img, (400, 300))
                downloaded_images.append(img_resized)
                image_names.append(f"Ïù¥ÎØ∏ÏßÄ {i}")

        if downloaded_images:
            print(f"\n‚úì Ï¥ù {len(downloaded_images)}Í∞úÏùò Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú ÏôÑÎ£å\n")
            rows = (len(downloaded_images) + 1) // 2
            display_images_grid(downloaded_images, image_names, rows=rows, cols=2)

    # ============================================
    # ÌîΩÏÖÄ Îç∞Ïù¥ÌÑ∞ Ï∂úÎ†• ÏòàÏ†ú
    # ============================================
    if img1 is not None:
        print("\n" + "="*60)
        print("üîç ÌîΩÏÖÄ Îç∞Ïù¥ÌÑ∞ ÏÉòÌîå Ï∂úÎ†•")
        print("="*60)

        # ÌäπÏ†ï ÏúÑÏπòÏùò ÌîΩÏÖÄ Í∞í Ï∂úÎ†•
        h, w = img1.shape[:2]
        center_y, center_x = h // 2, w // 2

        print(f"\nÏ§ëÏïô ÌîΩÏÖÄ ({center_x}, {center_y})Ïùò BGR Í∞í:")
        pixel = img1[center_y, center_x]
        print(f"  B: {pixel[0]}, G: {pixel[1]}, R: {pixel[2]}")

        # Ïù¥ÎØ∏ÏßÄ ÏùºÎ∂Ä ÏòÅÏó≠Ïùò ÌîΩÏÖÄ Í∞í Ï∂úÎ†• (5x5)
        print(f"\nÏ¢åÏ∏° ÏÉÅÎã® 5x5 ÏòÅÏó≠Ïùò ÌîΩÏÖÄ Í∞í (B Ï±ÑÎÑê):")
        print(img1[0:5, 0:5, 0])  # Blue Ï±ÑÎÑêÎßå

        # ÌÜµÍ≥Ñ Ï†ïÎ≥¥
        print(f"\nÏ±ÑÎÑêÎ≥Ñ ÌÜµÍ≥Ñ:")
        for i, color in enumerate(['Blue', 'Green', 'Red']):
            channel = img1[:, :, i]
            print(f"  {color} Ï±ÑÎÑê:")
            print(f"    ÌèâÍ∑†: {channel.mean():.2f}")
            print(f"    ÌëúÏ§ÄÌé∏Ï∞®: {channel.std():.2f}")
            print(f"    ÏµúÏÜåÍ∞í: {channel.min()}")
            print(f"    ÏµúÎåÄÍ∞í: {channel.max()}")

    print("\n" + "="*60)
    print("‚ú® ÌîÑÎ°úÍ∑∏Îû® Ï¢ÖÎ£å")
    print("="*60)


# ============================================
# ÏÇ¨Ïö©Ïûê Ï†ïÏùò URLÎ°ú Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìúÌïòÍ∏∞
# ============================================
def download_custom_image(url, display=True):
    """
    ÏÇ¨Ïö©ÏûêÍ∞Ä Ï†úÍ≥µÌïú URLÏóêÏÑú Ïù¥ÎØ∏ÏßÄÎ•º Îã§Ïö¥Î°úÎìúÌïòÍ≥† ÌëúÏãúÌïòÎäî Ìï®Ïàò

    ÏÇ¨Ïö© Ïòà:
        url = "https://example.com/image.jpg"
        img = download_custom_image(url)

    Args:
        url (str): Ïù¥ÎØ∏ÏßÄ URL
        display (bool): Ïù¥ÎØ∏ÏßÄÎ•º ÌëúÏãúÌï†ÏßÄ Ïó¨Î∂Ä

    Returns:
        numpy.ndarray: OpenCV Ïù¥ÎØ∏ÏßÄ Î∞∞Ïó¥
    """
    print(f"\nÏù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú ÏãúÏûë: {url}")
    img = download_image_to_memory(url)

    if img is not None and display:
        print_image_info(img)
        display_image(img, "Îã§Ïö¥Î°úÎìúÎêú Ïù¥ÎØ∏ÏßÄ")

    return img


# Ï∂îÍ∞Ä ÏÇ¨Ïö© ÏòàÏ†ú
"""
# ColabÏóêÏÑú ÏÇ¨Ïö©Ìï† Îïå:

# 1. Îã®Ïùº Ïù¥ÎØ∏ÏßÄ Îã§Ïö¥Î°úÎìú
url = "https://images.unsplash.com/photo-1514888286974-6c03e2ca1dba?w=800"
img = download_custom_image(url)

# 2. Ïù¥ÎØ∏ÏßÄ Ï†ïÎ≥¥ ÌôïÏù∏
print_image_info(img, "ÎÇ¥ Ïù¥ÎØ∏ÏßÄ")

# 3. Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
display_image(img, "Í∑ÄÏó¨Ïö¥ Í≥†ÏñëÏù¥")

# 4. Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨
processed = process_images(img)
display_image(processed['edges'], "Ïó£ÏßÄ Í≤ÄÏ∂ú Í≤∞Í≥º")

# 5. Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Î≥ÄÌôò Î∞è Ï†ÄÏû•
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('grayscale_image.jpg', gray)
"""
```
